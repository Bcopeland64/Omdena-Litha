```python
#import packages
import pandas as pd 
import numpy as np
from bertopic import BERTopic
```

```python
df = pd.read_json('/content/psychotherapy_transcripts.json')
```

```python
# create model 
 
model = BERTopic(verbose=True)
 
#convert to list 
docs = df.message.to_list()
 
topics, probabilities = model.fit_transform(docs)
```
